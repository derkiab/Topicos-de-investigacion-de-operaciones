```python
import numpy
import random
import math

M_dis=numpy.loadtxt("distancias.txt")
T=100000
T_final=0.001
beta=0.999

actual = numpy.arange(0,29)


while T>T_final:

  nuevo = actual.copy()
  n1, n2 = random.sample(range(29),2)
  nuevo[n1], nuevo[n2] = nuevo[n2], nuevo[n1]

  costo = 0
  for i in range (28): #1 menos pq buscamos otra wea
    a=actual[i]
    b=actual[i+1]
    costo=costo + M_dis[a][b]
  costo = costo + M_dis[actual[28]][actual[0]]


  #calcular costo solucion nueva

  costo_n = 0
  for i in range (28): #1 menos pq buscamos otra wea
    a=nuevo[i]
    b=nuevo[i+1]
    costo_n=costo_n + M_dis[a][b]
  costo_n = costo_n + M_dis[nuevo[28]][nuevo[0]]

  delta=costo_n - costo
  if costo_n < costo or random.random() < math.exp(-delta/T):
    actual=nuevo.copy()

  print(costo_n)
  T=T*beta
```
